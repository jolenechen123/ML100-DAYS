## 筆記
* 很多子圖
    * seaborn.heatmap : 將矩形數據繪製為顏色編碼矩陣
        * http://seaborn.pydata.org/generated/seaborn.heatmap.html
    * seaborn.PairGrid : 用於繪製數據集中成對關係的子圖網格
        * https://seaborn.pydata.org/generated/seaborn.PairGrid.html
* 合併圖
    * sns.kdeplot : 擬合併繪製單變量或雙變量核密度估計值
        *https://seaborn.pydata.org/generated/seaborn.kdeplot.html
* plt.scatter 散佈圖
    * https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
matrix1 = np.random.random((10,10))
matrix1
```

```python
matrix = (np.random.random((10,10)) - 0.5) * 2
matrix
```

```python
matrix = (np.random.random((10,10)) - 0.5) * 2

plt.figure(figsize=(10,10))

heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1., vmax = 1., annot = True)

plt.show()
```

```python
matrix = (np.random.random((1000,3)) - 0.5)*2
matrix
```

```python
nrow = 1000
ncol = 3
matrix = (np.random.random((1000,3)) - 0.5)*2

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2) #散佈圖
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
matrix = np.random.randn(nrow * ncol).reshape((nrow, ncol))

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2) #散佈圖
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
